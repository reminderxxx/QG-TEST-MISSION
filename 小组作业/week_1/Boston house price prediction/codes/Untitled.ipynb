5. 训练线性回归模型

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2 ,random_state=42)

model = LinearRegression()

model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"MSE = {np.sqrt(mse)}")
print(f"R2 = {r2}")
```

6. 手搓回归算法代码

```python
import numpy as np
from sklearn.model_selection import train_test_split

# 加载波士顿房价数据集
X = df.iloc[:, :-1].values  # 提取特征（除最后一列外的所有列）
Y = df.iloc[:, -1].values   # 提取目标变量（最后一列）

# 划分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 添加偏置项
X_b_train = np.c_[np.ones(X_train.shape[0]), X_train]
X_b_test = np.c_[np.ones(X_test.shape[0]), X_test]

# 最小二乘法实现线性回归
def least_squares(X_b, Y):
    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)
    return theta

# 训练模型
theta = least_squares(X_b_train, Y_train)

# 预测
Y_pred = X_b_test.dot(theta)

# 评估（均方误差）
mse = np.mean((Y_pred - Y_test) ** 2)
print(f"均方误差 (MSE): {mse}")
```

7. 最小二乘法

```python
import numpy as np

def least_squares(X, Y):
    # 添加偏置项
    X_b = np.c_[np.ones(X.shape[0]), X]
    
    # 计算参数 theta
    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)
    return theta
```

8. 梯度下降

```python
import numpy as np

def gradient_descent(X, Y, alpha=0.01, num_iters=1000):
    # 添加偏置项
    X_b = np.c_[np.ones(X.shape[0]), X]
    m, n = X_b.shape
    
    # 初始化参数
    theta = np.zeros(n)
    
    # 迭代更新参数
    for _ in range(num_iters):
        h = X_b.dot(theta)  # 预测值
        gradient = (1 / m) * X_b.T.dot(h - Y)  # 计算梯度
        theta -= alpha * gradient  # 更新参数
    return theta
```
